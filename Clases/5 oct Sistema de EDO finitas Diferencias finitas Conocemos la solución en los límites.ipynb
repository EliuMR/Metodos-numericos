## Conocemos la solución en los límites

Uso de la función edodif para 

$y''-y'+y-2e^{x}-3=0$, $y(0)=1$, $y(1)=5$ $0<=x<=1$

Forma estandar
$(2+h)y_{i-1}+(2h²-4)y_i+(2-h)y_{i+1}=4h²e^{x_i}+6h²$

$(P)y_{i-1}+(Q)y_i+(R)y_{i+1}=(S)$

```python
import matplotlib.pyplot as plt
from edodif import edodif
import numpy as np
def P(x,h):
    return 2+h
def Q(x,h):
    return 2*h**2-4
def R(x,h):
    return 2-h
def S(x,h):
    return 4*h**2*np.exp(x)+6*h**2
[u,v]=edodif(P,Q,R,S,0,1,1,5,20)
fig, ax=plt.subplots(1,1)
ax.plot(u,v,'ob')
ax.set_xlim(0,1)
ax.set_ylim(0,5)
```

```python
import matplotlib.pyplot as plt
from edodif import edodif
import numpy as np
def P(x,h):
    return 4+3*h
def Q(x,h):
    return -8
def R(x,h):
    return 4-3*h
def S(x,h):
    return -4*h**2*x+10*h**2
[u,v]=edodif(P,Q,R,S,1,2,3,4,20)
fig, ax=plt.subplots(1,1)
ax.plot(u,v,'o')
```

### Ecuaciones diferenciales ordinarias con codiciones en los bordes de las derivadas

$y''-y'+y-2e^{x}-3=0$, 

$y'(0)=0.5$, $y(1)=5$ $0<=x<=1$

Forma estandar
$(2+h)y_{i-1}+(2h²-4)y_i+(2-h)y_{i+1}=4h²e^{x_i}+6h²$

$(P)y_{i-1}+(Q)y_i+(R)y_{i+1}=(S)$

```python
from edodifdi import edodifdi
import numpy as np
import matplotlib.pyplot as plt


def P(x, h):
    return 2 + h


def Q(x, h):
    return 2 * h**2 - 4


def R(x, h):
    return 2 - h


def S(x, h):
    return 4 * h**2 * np.exp(x) + 6 * h**2


[u, v] = edodifdi(P, Q, R, S, 0, 0.5, 1, 5, 20)
fig, ax = plt.subplots(1, 1)
ax.plot(u, v, 'o')
```
