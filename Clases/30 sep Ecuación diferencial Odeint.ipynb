$y''-y'-x+y+1=0$  , y(0)=1 , y'(0)=2

$z=y'$

$z'-z-x+y+1=0$

$y'=f(x,y,z)=z, y(0)=1$

$z'=x-y+z-1,z(0)=2$

```python
from scipy import integrate
import matplotlib.pyplot as plt
import numpy as np
def f(yz0,x):
    y,z=yz0
    return[z,x-y+z-1]
yz0=[1,2]
x=np.linspace(0,10,200)
sol=integrate.odeint(f,yz0,x)
fig, ax=plt.subplots(1,1,figsize=(6,4))
ax.plot(t,sol[:,0],'r-.',label='y(x)')
ax.plot(t,sol[:,1],'b-.',label='dy(x)/dx')
ax.set_xlabel('t')
plt.legend()
```

$y(0)=1.5$ $y'(0)=4$

```python
from scipy import integrate
import matplotlib.pyplot as plt
import numpy as np
import math
def fr(yz0,x):
    y,z=yz0
    return[z,z+math.sin(x)-y-1]
yz0=[1.5,2.5]
x=np.linspace(0,5,200)
so=integrate.odeint(fr,yz0,x)
fig, ax=plt.subplots(1,1,figsize=(6,4))
ax.plot(t,so[:,0],'r-.',label='y(x)')
ax.plot(t,so[:,1],'b-.',label='dy(x)/dx')
ax.set_xlabel('t')
plt.legend()
```

```python
from scipy import integrate
import matplotlib.pyplot as plt
import numpy as np
import math
m=1
mu=4.7
k=.9
def fr(xz0,t):
    x,z=xz0
    return[z,-(mu*(x**2-1)*z+k*x)/m]
xz0=[0.75,0]
t=np.linspace(0,40,200)
so=integrate.odeint(fr,xz0,t)
fig, ax=plt.subplots(1,1,figsize=(6,4))
ax.plot(t,so[:,0],'r',label='x(t)')
#ax.plot(t,so[:,1],'b-.',label='dx(t)/dt')
ax.set_xlabel('t')
plt.legend()
```
