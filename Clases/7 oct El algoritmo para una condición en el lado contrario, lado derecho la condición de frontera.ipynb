Un problema importante en ingenieria es la defexión de una viga de sección transversal uniforme sujeta a una carga distribuisa y colocada con sus extremos apoyados de tal formma que no pueda deflectarse
LA ecuación que describe esto:$\frac{d²w}{dx²}=\frac{S}{El}w+\frac{qx}{2El}(x-l)$
    
w-deflexión desde el lado izquierdo

q-intensidad de carga por unidad de longitud

S-tensión en los putos extremos

I-momento principal de inercia

L-longitud de la viga

La condición de no deflexión en los extremos se expresa con: w(0)=w(L)=0
    
Suponga una viga de acero con la siguientes características:
    L=10m , 1=14060N/m,E=20.7x10⁷kPa,S=4.4kN,I=5m²


Calcule la deflexión.

$w''=\frac{S}{EI}+\frac{qx}{2EI}(x-L)$

$\frac{w_{i+1}-2w_i+w_{i-1}}{h²}=\frac{S}{EI}+\frac{qx}{2EI}(x-L)$

$w_{i+1}-2w_i+w_{i-1}=h²\frac{S}{EI}+h\frac{qx}{2EI}(x-L)$

$w_{i+1}+(-2-\frac{h²S}{EI})w_i+w_{i-1}=h²\frac{qx_i}{2EI}(x_i-L)$

```python
from edodifdd import edodifdd
from edodif import edodif
import numpy as np
import matplotlib.pyplot as plt 
Sa=4.4e3
Ia=5
E=20.7e10
q=1460
L=10
def P(x,h):
    return 1
def Q(x,h):
    return -2-h**2*Sa/(E*Ia)
def R(x,h):
    return 1
def S(x,h):
    return h**2*q*x*(x-L)/(2*E*Ia)
[u,v]=edodif(P,Q,R,S,0,0,10,0,1000)
fig, ax=plt.subplots(1,1)
ax.plot(u,v,'og')
```

### Método de diferencias finitas para resolver ecuaciones diferenciales parciales.

```python
#Metodo explícito de diferencias finitas para una EDP parabólica
#U(i,j+1)=(P)U(i-1,j)+(Q)U(i,j)+RU(i+1,j)
def edpdif(P,Q,R,U,m):
    u=[U[0]]
    for i in range(1,m):
        u=u+[P*U[i-1]+Q*U[i]+R*U[i+1]]    #Calculo de puntos
    u=u+[U[m]]
    return u
from pylab import*
m=15
n=100
Ta=60; Tb=40
To=25
dx=0.1;dt=0.01
L=1
k=4
U=[Ta]
for i in range(1,m):
    U=U+[To]
U=U+[Tb]
lamb=dt/(k*dx**2)
P=lamb
Q=1-2*lamb
R=lamb
title('Curvas de ditribución térmica')
xlabel('X(distancia)')
ylabel('U(temperatura)')
x=[0]
for i in range(1,m+1):
    x=x+[i*dx]
plot(x,U,'or')
for j in range(n):
    U=edpdif(P,Q,R,U,m)
    if j%10==0:
        plot(x,U,'-r');
        plot(x,U,'.r')
grid(True)
show()
```
