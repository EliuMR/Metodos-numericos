Determine la slución de la ecuación:

$xe^{-x}=0.2$

```python
from biseccion import bisectionmethod
import sympy
from scipy import optimize
import math
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x*np.exp(-x)-0.2
a=np.linspace(0,8)
b=f(a)
bisectionmethod(f,0,1,100,0.001)
bisectionmethod(f,2,3,100,0.001)
plt.plot(a,b)
plt.axhline(y=0,color='black')
plt.grid()
plt.show()
xc=optimize.bisect(f,0,0.5)
print(xc)
```

Mínimo de la función

$f(x)=x³-12x²+40.25x-36.5$

intervalo de 3 a 8

```python
from biseccion import bisectionmethod
import math
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**3-12*x**2+40.25*x-36.5
def f1(x):
    return 3*x**2-24*x+40.25
a=np.linspace(1,8)
b=f(a)
bisectionmethod(f1,3,7,100,0.001)

plt.plot(a,b)
plt.axvline(x=3,color='blue')
plt.axvline(x=8,color='blue')
plt.axhline(y=0,color='black')
plt.grid()
plt.show()
```

Integrar

$\int_0^8(xe^{-x^{0.8}}+0.2)dx$

```python
import numpy as np
from scipy import integrate
def f(x):
    return x*np.exp(-x**0.8)+0.2
a=0
b=8
x=np.linspace(a,b,1000)
y=f(x)
r=integrate.trapz(y,x)
s=integrate.simps(y,x)
plt.plot(x,y)
plt.grid()
plt.show()
print(r,s)
```

Resolver:
    
$\frac{dy}{dt}=\frac{t³-2y}{t}$

para 1<t<3 con y=4.2 in t=1

```python
import sympy
from scipy import integrate
t=sympy.symbols('t')
y=sympy.symbols('y')
f=(t**3-2*y(t))/t
f_np=sympy.lambdiy((y(t),t),f)
t=np.linspace(1,3,100)
y0=4.2
y_t=integrate.odeint(f_np,y0,t)
plt.plot(t,y_t)
```
