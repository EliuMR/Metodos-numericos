# Agrupación de datos

```python
import numpy as np
import pandas as pd
```

```python
gender=['Male','Female']
es=['Bajo','Medio','Alto']
```

```python
n=500

gender_data=[]
income_data=[]
for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(es))
```

```python
height=160+30*np.random.randn(n)
weight=65+25*np.random.randn(n)
age=30+12*np.random.randn(n)
income=18000+3500*np.random.randn(n)
```

```python
dic={'Genero':gender_data,'Estatus':income_data,'Altura':height,'Peso':weight,'Edad':age,'Ingesos':income}
```

```python
data=pd.DataFrame(dic)
```
