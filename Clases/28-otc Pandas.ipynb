```python
#Chido
Patriots_Blackman=[11.5,10.85,11.15,10.7,11.1,11.6,11.85,11.1,10.95,10.5]
Colts_Blackman=[12.7,12.75,12.5,12.55]
Patriots_Prioleau=[11.8,11.2,11.5,11.0,11.45,11.95,12.30,11.55,11.35,10.9]
Colts_Prioleau=[12.35,12.3,12.95,12.15]
```

```python
A1=np.concatenate((Patriots_Blackman,Colts_Blackman),axis=None)
```

```python
A2=np.concatenate((Patriots_Prioleau,Colts_Prioleau),axis=None)
```

```python
dic={'Blackman':A1,'Prioleau':A2}
```

```python
data=pd.DataFrame(dic)
```

```python
data
```

```python
data.plot(kind='scatter',x='Blackman',y='Prioleau')
```

```python
#regrecin lineal
import scipy.stats
slope,intercept,r_value,p_value,std_err=\
scipy.stats.linregress(A1,A2)
print('r=',r_value)
```

```python
import matplotlib.pyplot as plt
data2=data[0:10]
data2.plot(kind='scatter',x='Blackman',y='Prioleau')
x=np.linspace(10.5,12.,100)
print(data2)
slope,intercept,r_value,p_value,std_err=\
scipy.stats.linregress(data2['Blackman'],data2['Prioleau'])
print('r=',r_value)
print(intercept)
print(slope)
y=intercept+slope*x
plt.plot(x,y)
```
