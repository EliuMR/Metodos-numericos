```
#importamos las librerias
import matplotlib.pyplot as plt
import scipy as sc
import numpy as np
 
#creamos la función a encontar mínimos, notado que el parámetro xy esun vector xy=[x,y]
func = lambda xy: np.sin(1/2*xy[0]**2-1/4*xy[1]**2+3)*np.cos(2*xy[0]+1-np.e**xy[1])
 
#Creamos nuestro espacio donde graficaremos, colocando a Z como ceros para después colocar los valores obtenidos en la derecha
res=100
_X=np.linspace(-2,2,res)
_Y=np.linspace(-2,2,res)
_Z=np.zeros((res,res))
 
#Obtenemos los valores de la matriz Z con la función y o vectores _x,_y
for ix,x in enumerate(_X):
  for iy,y in enumerate(_Y):
    _Z[iy,ix]=func([x,y])
 
#Creamos nuestro punto aleatorio, dentro de nuestro plano xy, desde donde buscaremos el mínimo y graficamos
p_0=np.random.rand(2)*4-2
print('Punto inicial',p_0)
plt.plot(p_0[0],p_0[1],'o',c='r')
 
#GRaficamos nuestra función, pero como superficies de nivel para que se vea más claro el recorrido
plt.contourf(_X,_Y,_Z,10)
plt.colorbar()
 
#Radio de aprendizaje
lr=0.01
 
#Creamoos nuestro vector gradiente, en espe caso de dos dimensiones
gradiente=np.zeros(2)
 
#Tamaño de nuestra variación y tener la derivada
h=0.001
 
#Realizamos nuestro método iterativo,donde calculamos el gradiente en cada punto que nos encontramos p_0 y nos movemos a p_0-lr*\nabla f
#Además obtenemos como una aproximación a lo que es la derivada parcial, por eso elegimos a h<<1
#Primero con un ciclo for para que lo haga 5000 veces y con otro ciclo for para que haga la derivada de cada componenete de p_0 es decir con respecto a x e y
_T=np.copy(p_0)
for i in range(5000):
  for it, xy in enumerate(p_0):
    _T=np.copy(p_0)
    _T[it]=_T[it]+h
    deriv=(func(_T)-func(p_0))/h
    gradiente[it]=deriv
  p_0=p_0-lr*gradiente
 
#Graficamos cada 100 puntos y veamos hacia dónde se dirige nuestro punto
  if(i%100==0):
    plt.plot(p_0[0],p_0[1],'.',c='black')
  if(i==4999):
    plt.plot(p_0[0],p_0[1],'.',c='white')
    print('Punto final final',p_0)

 
plt.show()
```
