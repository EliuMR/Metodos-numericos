```
class Cliente:
  def __init__(self,nombre,cantidad,operacion):
    self.nombre=nombre
    self.cantidad=cantidad
    self.operacion=operacion
  def transaccion(self):
    if self.operacion=='retiro':
      self.cantidad=-self.cantidad


class Banco(Cliente):
  def estado(self):
    self.cuenta=0
    self.cuenta=self.cuenta+self.cantidad

  def recibo(self):
    print('Nombre: ',self.nombre)
    print('Su operación fue: ',self.operacion, ', la cantidad de:',self.cantidad)
    print('El estado de cheque es: ',self.cuenta)

cliente1=Banco('Roberto',1500,'retiro')
cliente1.transaccion()
cliente1.estado()
cliente1.recibo()

```

```
class triangulo:
  def __init__(self):
    self.lado1=int(input('Ingrese valor de lado 1: '))
    self.lado2=int(input('Ingrese valor de lado 2: '))
    self.lado3=int(input('Ingrese valor de lado 3: '))
  def tipo_triangulo(self):
    if self.lado1<self.lado2+self.lado3 and self.lado2<self.lado3+self.lado1 and self.lado3<self.lado1+self.lado2:
      if self.lado1==self.lado2==self.lado3:
        print ('El triangulo es equilatero')
      elif self.lado1==self.lado2 or self.lado2==self.lado3 or self.lado1==self.lado3:
        print('El triangulo es isosceles')
      elif (self.lado1)**2==(self.lado2)**2+(self.lado3)**2 or (self.lado2)**2==(self.lado2)**2+(self.lado1)**2 or (self.lado3)**2==(self.lado1)**2+(self.lado2)**2:
        print('Es un triángulo rectángulo')
      else:
        print('El triángulo es escaleno')
    else: 
      print('Con las medidas dadas no se pude formar un triángulo')
figura=triangulo()
figura.tipo_triangulo()
```

```
class Agenda:
  def __init__(self):
    self.contactos=[]
  def menu(self):
    MENU=[['Agenda personal'],['1.Añadir'],['2.Lista de contactos'],
          ['3.Buscar contacto'],['4.Editar contacto'],['5.Cerrar agenda']]
    
    for i in range(6):
      print(MENU[i][0])
    
    opcion=int (input('Introduzca la opción deseada: '))
    if opcion==1:
      self.añadir()
    elif opcion==2:
      self.lista()
    elif opcion==3:
      self.buscar()
    elif opcion==4:
      self.editar()
    elif opcion==5:
      print('Saliendo de la agenda')
      exit()
    
    self.menu()

  def añadir(self):
    print('----------------')
    print('Añadir contactos')
    print('----------------')
    nom=input('Introduzca el nombre: ')
    telf=input('Introduzca el numero: ')
    email=input('Introdusca el e-mail: ')
    self.contactos.append({'Nombre':nom,'Telefono':telf,'email':email})
  
  def lista(self):
    print('------------------')
    print('Lista de contactos')
    print('------------------')
    if len(self.contactos)==0:
      print('No existen contactos')
    else:
      for i in range(len(self.contactos)):
        print(self.contactos[i]['Nombre'])
  
  def buscar(self):
    print('---------------')
    print('Buscar contacto')
    print('---------------')
    name=input('Introduzca el contacto: ')
    for i in range(len(self.contactos)):
      if name==self.contactos[i]['Nombre']:
        print('--Datos del contacto--')
        print('Nombre',self.contactos[i]['Nombre'])
        print('Telefono',self.contactos[i]['Telefono'])
        print('e-mail',self.contactos[i]['email'])
        return i
    for i in range(len(self.contactos)) :  
      if not name==self.contactos[i]['Nombre']:
        print ('El contacto no existe')

  def editar(self):
    print('----------------')
    print('Editar contactos')
    print('----------------')
    data=self.buscar()
    bandera=False
    while bandera==False:
      print('Seleccione la opción a editar')
      print('1 Nombre')
      print('2 Telefono')
      print('3 E-mail')
      print('4 Volver')
      opcion=int(input(' '))
      if opcion==1:
        nom=input('Escriba el nuevo nombre: ')
        self.contactos[data]['Nombre']=nom
      elif opcion==2:
        tel=input('Escribe el nuevo numero: ')
        self.contactos[data]['Telefono']=tel
      elif opcion==3:
        ema=input('Escrib el nuevo e-mail: ')
      elif opcion==4:
        bandera==True
  
agenda=Agenda()
agenda.menu()


  
```
