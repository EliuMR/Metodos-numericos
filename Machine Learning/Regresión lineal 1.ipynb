```
#Importamos librerias
import numpy as np
import matplotlib.pyplot as plt

# Escribimos nuestros datos de las calificaciones
Matematicas=np.array([6,4,8,5,3.5,9])
Fisica=np.array([6.5,4.5,7,5,4,8])
 
#Ahora escribimos las matrices X,Y recordando que la matriz X, debe tener una columna de unos
X=np.array([[1,6],[1,4],[1,8],[1,5],[1,3.5],[1,9]])
Y=np.array([[6.5],[4.5],[7],[5],[4],[8]])
 
#Encontramos la matriz W, que nos da los valores que nos harán mína la función de coste, al ser un caso bidimensional, nos dará la ordenada al origen así como la pendiente
W=np.linalg.inv(X.T @ X) @ X.T @ Y

#Pasamos a dibujar la recta en el espacio de (0,10) en en el eje "x" y utilizando los valores obtenidos en W, obtenemos "y" con la ecuación de la recta y=mx+b
plt.plot([0,10],[W[0]+W[1]*0,W[0]+W[1]*10],c='r')
plt.plot(Matematicas,Fisica,'o')
plt.title('Regresión lineal')
plt.xlabel('Matemáticas')
plt.ylabel('Física')
plt.grid()
plt.show()
```
