```
import matplotlib.pyplot as plt
import numpy as np
 
a=(0,0,1,1)
b=(0,1,0,1)
plt.text(0, 0, u'0', fontsize = 18, horizontalalignment='center', verticalalignment='center',color='red') 
plt.text(0, 1, u'0', fontsize = 18, horizontalalignment='center', verticalalignment='center',color='red') 
plt.text(1, 0, u'0', fontsize = 18, horizontalalignment='center', verticalalignment='center',color='red') 
plt.text(1, 1, u'1', fontsize = 18, horizontalalignment='center', verticalalignment='center',color='green') 
plt.plot(a,b,c='w')
def f(x):
  return (6-4*x)/5
s=np.linspace(.2,1.01,10)
plt.plot(s,f(s))
plt.grid()
plt.axes().set_aspect('equal')
plt.xlabel('Café')
plt.ylabel('Azúcar')
plt.savefig('Puerta and')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
def f(x):
  if x<=0:
    return 0
  else: return 1
y=np.zeros(1000)
x=np.linspace(-5,5,1000)
for i in range(len(x)):
  y[i]=f(x[i])
plt.grid()
plt.savefig('Escalonada')
plt.plot(x,y,c='r')
```

```
def g(x):
  return 1/(1+np.e**(-x))
plt.grid()
plt.savefig('Sigmoide')
plt.plot(x,g(x),c='g')
```

```
def h(x):
  return np.tanh(x)
plt.plot(x,h(x))
plt.grid()
plt.savefig('Tanh')
```

```
def j(x):
  if x<0:
    return 0
  else: return x
x=np.linspace(-5,5,1000)
l=np.zeros(1000)
for i in range(len(x)):
  l[i]=j(x[i])
plt.plot(x,l,c='purple')
plt.grid()
plt.savefig('Relu')
```
