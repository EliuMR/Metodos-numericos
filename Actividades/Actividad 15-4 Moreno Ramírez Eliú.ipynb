Un parachoques de seguridad se coloca al final de un
pista de carreras para detener los coches fuera de control. 
El parachoques está diseñado de tal manera que la fuerza que
el parachoques se aplica al coche es una función
de la velocidad v y el desplazamiento x de
el borde delantero del parachoques según la ecuación:

$F=Kv^3(x+1)³$

donde $K = 30 (s kg) / m^5$ es una constante.
Un automóvil con una masa m de 1500 kg golpea el parachoques a una velocidad de 90 km / h.
Determine y grafique la velocidad del automóvil en función de su posición para $0<=x<=3$
metro.

```python
import numpy as np
import matplotlib.pyplot as plt
#Utilizando el método de rugekutta
#x,y punto inicial
#Iteraciones
def rugekutta(f,x,y,h,m):
    u=[]
    v=[]
    for i in range(m):
        k1=h*f(x,y)
        k2=h*f(x+h/2.0,y+k1/2.0)
        k3=h*f(x+h/2.0,y+k2/2.0)
        k4=h*f(x+h,y+k3)
        y=y+1.0*1/6*(k1+2*k2+2*k3+k4)
        x=x+h
        u=u+[x]
        v=v+[y]
    return [u,v]
K=30 #(skg/m⁵)
m=1500 #kg
v0=90*1000/3600 #m/s
def f(v,x):
    return -(K*v**2*(x+1)**3)/m
[u,v]=rugekutta(f,0,v0,0.01,300)

fig, ax=plt.subplots(1,1,figsize=(8,6))
ax.plot(u,v,'r',label='velocidad(m/s)')

plt.legend()

```
