Utilice los datos para calcular el caudal y luego integre el caudal para obtener un
estimación de la cantidad total de agua que fluye en el río durante un año.

El caudal, Q (volumen de agua por segundo), en cada punto de datos se obtiene mediante
multiplicar la velocidad del agua por el ancho y la altura del área de la sección transversal de
el agua que fluye en el canal:
    $Q=vwh (m³/s)$
La cantidad total del flujo del agua es estimada por la integral:

$V=60*60*24\int_{t_1}^{t_2}Qdt$

El caudal se da en metros cúbicos por segundo, lo que significa que el tiempo debe tiener unidades de segundos. Dado que los datos se dan en términos de días, la integral es múltiple
multiplicado por ($60*60*24)s / día.$

```python
import matplotlib.pyplot as plt
#Utilizando el método de simpson discreto
def simpsonDiscreto(f,a,b,m):
    h=1.0*(b-a)/m
    s=0.0
    for i in range(1,m):
        s=s+2*(i%2+1)*f[i]
    s=h/3*(f[0]+s+f[m])
    return s
d=f[0]
r=60*60*24*simpsonDiscreto(Q,1,366,12)
plt.plot(d,Q)
plt.grid()
plt.show()
print('La cantidad de agua que fluye en el río durante un año es: ',r,'m³')
```
