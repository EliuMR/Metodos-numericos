Calcular la siguiente integral: $S=\int_R\int x^{2}\sqrt{9-y^2}dA$ Donde R es la región $x^{2}+y^{2}=9$. Usar m=4 en ambas direcciones.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import sympy
def simpson(f,a,b,m):
    h=(b-a)/float(m)
    s=0
    x=a
    for i in range (1,m):
        s=s+2*(i%2+1)*f(x+i*h)
    s=h/3*(f(a)+s+f(b))
    return s
def simpson3lv(f,ax,bx,h,q,mx,my):
    s=sympy.Symbol('y')
    dx=(bx-ax)/float(mx)
    v=ax
    r=[]
    for i in range(0,mx+1):
        def g(y):
            return f(v,y)
        w1=h(v)
        w2=q(v)
        #print ('i,w1,w1',i,w1,w2)
        u=simpson(g,w1,w2,my)
        r=r+[u]
        v=v+dx
    s=0
    for i in range(1,mx):
        s=s+2*(2-(i+1)%2)*r[i]
    s=dx/3.0*(r[0]+s+r[mx])
    return s



def f(x,y):
    return x**2*math.sqrt(9-y**2)
def h(x):
    return -math.sqrt(9-x**2)
def g(x):
    return math.sqrt(9-x**2)
r=simpson3lv(f,-3,3,h,g,4,4)

x1=np.linspace(-3,3,100)
f=(9-x1**2)**(1/2)
g=-(9-x1**2)**(1/2)
fig, ax=plt.subplots(1,1,figsize=(5,5))
ax.plot(x1,f)
ax.plot(x1,g)
ax.set_ylim(-4,4)
ax.set_xlim(-4,4)
ax.grid()
plt.show()
print ('El resultado de la integral es',r)
```

```python
import numpy as np
def simpson3Discreto(f,ax,bx,ay,by,mx,my):
    hx=(bx-ax)/(mx)
    hy=(by-ay)/(my)
    r=np.zeros(my+1)
    for i in range(0,my+1):
        for j in range(1,mx):
            #print(2*(j%2+1))
            r[j]=r[j]+2*(j%2+1)*f[j,i]
        r[i]=hy/3.0*(f[0,i]+r[i]+f[mx,i])
        #print(i,f[0,i],f[mx,i])
    s=0
    #print(hy,hx)
    for i in range(1,my):
        #print(2*(2-(i+1)%2))
        s=s+2*(2-(i+1)%2)*r[i]
    s=hx/3.0*(r[0]+s+r[my])
    return s
f=np.array([[0.04,0.08,0.12,0.16,0.20],
            [0.41,0.47,0.53,0.58,0.62],
            [0.81,0.83,0.84,0.83,0.82],
            [0.76,0.70,0.62,0.53,0.42],
            [0.06,0.02,0.01,0.01,0.02]])
ax=0.2
bx=1.0
ay=0.1
by=0.5
m=4
n=4
volumen=simpson3Discreto(f,ax,bx,ay,by,m,n)
print ('El volumen es :',volumen)
```
