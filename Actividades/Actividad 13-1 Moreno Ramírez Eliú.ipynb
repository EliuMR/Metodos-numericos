Ejercicio 1. Crea un programa que pida al usuario 10 números enteros, los guarde en un array y luego le pregunte de forma repetitiva qué número quiere buscar. Le responderá si dicho número estaba o no entre los datos que se habían introducido inicialmente. Dejará de repetirse cuando se introduzca el número 0.

```python
import numpy as np
n=np.zeros(10)
for i in range(10):
    s=int(input('Ingrese su número'))
    n[i]=s
buscar=int(input('Cuál es el número que desea buscar, (si presiona 0 significa que no desea buscar ningún número)'))
while not buscar==0:
    x=0
    for i in range(10):
        if buscar==n[i]:
            print('El número',buscar, ' se encuentra en la localización: ', i+1)
        else:
            x=x+1
        if x==10:
            print('El número',buscar ,' no se encuentra en el array')
    buscar=int(input('Desea buscar otro número si es así,cuál es el número que desea buscar, (si presiona 0 significa que no desea buscar ningún número)'))
print('Finalizado')
```
