La NASA lo contrató para desarrollar un breve script de Python que cuando se ejecuta, le pide al usuario que ingrese su peso en la tierra y luego seleccione el planeta que está ocupando actualmente. A continuación, el programa debe calcular e imprimir en la pantalla el peso del individuo en su planeta actual. Utilice una tabla de peso relativo como la que se reproduce a continuación para simplificar el cálculo. Tres notas: (1) la función input () en Python devuelve una cadena, que deberá convertir en un entero usando la función int (), (2) es más fácil pedirle al usuario que ingrese un número para el planeta selección (es decir, ingrese 1 para Mercurio, 2 para Venus, etc.), y (3) debe verificar la selección del planeta usando una declaración "if" para asegurarse de que sea válida.

```python
planetas=[0.38,0.91,1.0,0.38,2.34,1.06,0.92,1.19]
peso=int(input('Introduzca su peso en la tierra en kilogramos '))
num=int(input('Escriba el número de acuerdo al planeta Mercurio(1), Venus(2), Tierra(3), Marte(4), Jupiter(5), Saturno(6), Urano(7), Neptuno (8)'))
s=input('¿Está seguro que ese es el planeta? Escriba "y" para sí y "n" para no ')
if s=="y":
    w=peso*planetas[num-1]
    print ('Su peso en el planeta actual es: ',w)
else:
    print ('Reinicie el código')
```
