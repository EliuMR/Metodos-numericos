The idela gas equation relates the volume (V in L), temperature (T in K), pressure (P in atm), and the amount of gas (number of moles n) by: $p=\frac{nRT}{V}$ where R=0.08206(L atm)/(mol K)

The van der Waals equation gives the relationship between these quantities for a real gas by:
    
$(P+\frac{n²a}{V²})(V-nb)=nRT$

where a and b are constants
Calcule el volumen of 2 mol $CO_2$ at temperature of 50°C, and pressure of 6atm For $CO_2$,a=3.59(L² atm)/(mol²) and b=0.0427 L/mol.

```python
#Utilizando el método de la bissección
import numpy as np
import matplotlib.pyplot as plt

n=2 #mol
T=323.15 #k ó 50°C
a=3.59 #(L² atm)/(mol²) 
b=0.0427 #L/mol
R=0.08206 #(L atm)/(mol k)
P=6 #atm
def vanderW(v):
    return (P+(n**2*a)/v**2)*(v-n*b)-n*R*T
x=np.linspace(0,15,100)
plt.axhline(y=0,color='black')
plt.axvline(x=8.65,color='red')
y=vanderW(x)
plt.grid()
plt.xlabel('V')
plt.title('Ecuacion de Van der Waals')
plt.plot(x,y)
plt.show()
```

```python
#Vemos que la raíz está entre 8 y 10
#Utilizando el método de bisección
import numpy as np
def bisectionmethod(h,a,b,maxIter,TOL):
    stop=False
    fa=h(a)
    fb=h(b)
    
    #check starting values
    if(fa*fb>=0):
        print ('error')
    else:
        #Bisection iteration
        for i in range(maxIter):
            c=(a+b)/2
            fc=h(c)
            #print("c=",c,"and f(c)",fc)
            
            #check for convergence
            if(np.fabs(fc)<TOL):
                return c
                #print("Root found at",c)
                stop=True
                break
            #if not conerged, determine end point
            #to replace
            if(fa*fc<0):
                b=c
                fb=fc
            else:
                a=c
                fa=fc
        if stop == False:
            print("Reached maxIter, currently at",c)
volumen=bisectionmethod(vanderW,8,10,100,0.0001)
print('El volumen es:',volumen)
```
