Resuelva el siguiente sistema de ecuaciones:

$2x_1+3x_2=8$

$x_1-2x_2=-10$

```python
import scipy as sc
from scipy import linalg
import numpy as np
A=sc.matrix([[2,3],[1,-2]])
B=np.array([8,-10])
C=linalg.solve(A,B)
print ('Para x_1=',C[0],'Para x_2=',C[1])
```
