13. Calcule la integral $A=\int_1 ^{\infty}x^{-2}e^{-x^2}dx$. Use la regla de Simpson.

```python
#Cambio de variable propuesto x=1/t
import math
def simpson(f,a,b,m):
    h=(b-a)/float(m)
    s=0
    x=a
    for i in range (1,m):
        s=s+2*(i%2+1)*f(x+i*h)
    s=h/3*(f(a)+s+f(b))
    return s
def a(t):
    return -math.exp(-(1/t**2))
#Se escogió un valor muy cercano a 0 para que se pudiera hacer la integral
r=simpson(a,1,0.00001,20)
print('La integral es :',r)
```

18 La función $\Gamma(\alpha)=\int_0^{\infty}x^{\alpha-1}e^{-x}dx$ se denomina función Gamma y es usada en algunos modelos de probabilidad . Encuentre un valos aproximado de $\Gamma(2)$
