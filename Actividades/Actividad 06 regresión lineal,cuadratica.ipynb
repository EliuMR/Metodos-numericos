Se ha sugerido que el aumento de la temperatura media global se debe a
la disminución del número de piratas. También se le ha pedido que realice
un análisis de regresión lineal y cuadrática sobre el número aproximado de
piratas en el mundo (y) versus año (x). 

```python
import numpy as np
import numpy.linalg as nl
import matplotlib.pyplot as plt
n=13
x=np.array([1890,1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010])
y=np.array([13.8,13.7,13.7,13.6,13.8,13.9,14,14,14,14,14.2,14.4,14.6])
if ((n!=x.size) or (n!=y.size)):
    print ("Error")
Xsum = np.sum(x)
Ysum = np.sum(y)
XYsum = np.sum(np.dot(x,y))
X2sum = np.sum(np.dot(x,x))
x2=np.multiply(x,x)
x3=np.multiply(x2,x)
X3sum = np.sum(np.dot(x2,x))
X4sum = np.sum(np.dot(x3,x))
X2Ysum = np.sum(np.dot(x2,y))
A = np.array([[n, Xsum],[Xsum,X2sum]])
f=np.array([Ysum,XYsum])
c=nl.solve(A,f)
A1 = np.array([[n, Xsum,X2sum],[Xsum,X2sum,X3sum],[X2sum,X3sum,X4sum]])
f1=np.array([Ysum,XYsum,X2Ysum])
c1=nl.solve(A1,f1)

yLR = c[0]+c[1]*x
yLC = c1[0]+c1[1]*x+c1[2]*x**2

plt.plot(x,y,'o',x,yLR)
plt.plot(x,yLC)
plt.xlabel ('Año')
plt.ylabel ('Temperatura')
plt.show()
```

a + b ⋅ x + c ⋅ x^2 + d ⋅ x^3)$ análisis de regresión de la capacidad calorífica (y) versus
temperatura (x). Uso de la función curve_fit () del Scipy
biblioteca es aceptable para el editor. Comente sobre la calidad del ajuste
para las distintas curvas de regresión.
