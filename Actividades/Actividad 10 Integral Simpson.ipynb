La siguiente función se denomina función error: $erf(x)=\frac{2}{\sqrt{\pi}}\int_0 ^x e^{-t^2}dt$ . Esta función se usa para definir la función densidad de la probabilidad de la distribución estándar. Calcular $erf(1)$ con la fórmula de Simpson.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
def simpson(f,a,b,m):
    h=(b-a)/float(m)
    s=0
    x=a
    for i in range (1,m):
        s=s+2*(i%2+1)*f(x+i*h)
    s=h/3*(f(a)+s+f(b))
    return s
def g(t):
    return 2/(math.sqrt(math.pi))*math.exp(-t**2)
a=0.0
b=1.0
m=int(input('Ingrese el número de intervalos que desea, que sea par.'))
r=simpson(g,a,b,m)
x=np.linspace(a,b,20)
y=np.linspace(g(a),g(b),20)
plt.plot(x,y)
plt.show()
print('La integral por el método de Simpson es; ',r)
```

Una empresa está vendiendo una licencia para manejo de un nuevo punto de venta. La experiencia indica que dentro de t años, la licencia generará utilidades según $f(t)=1400+490t$ dolares por años. Si la tasa de interés r permanece fija durante los próximos n años, entnces el valo presente de la licencia se puede calcular con $V(n)= \int_o ^n f(t) e^{rt}dt$. Calcule $V$ si $n=5 años$ y $r=0.07$. Use la fórmula trapecio y de Simpson.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def trapecio(f,a,b,m):
    h=(b-a)/float(m)
    s=0.0
    for i in range (1,m):
        s=s+f(a+i*h)
    r=h/2.0*(f(a)+2*s+f(b))
    return r
def simpson(f,a,b,m):
    h=(b-a)/float(m)
    s=0
    x=a
    for i in range (1,m):
        s=s+2*(i%2+1)*f(x+i*h)
    s=h/3*(f(a)+s+f(b))
    return s
r=0.07
b=5
a=0
m=int(input('Eliga m, es decir el número de subintervalos que desee, que sea par.'))
def g(t):
    return (1400+490*t)*math.exp(r*t)
re1=trapecio(g,a,b,m)
re2=simpson(g,a,b,m)
x=np.linspace(a,b,100)
y=np.linspace(g(a),g(b),100)
plt.semilogx(x,y)

plt.show()
print('El resultado de la integral por el método del trapecio es: ',re1)
print('El resultado de la integral por el método de simpson: ',re2)
```
