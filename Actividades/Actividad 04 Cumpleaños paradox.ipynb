Su informe final al departamento como consultor debe consistir de una discusión de sus hallazgos y, lo más importante, una gráfica de la probabilidad de que una clase contenga dos estudiantes con el mismo cumpleaños versus el tamaño de la clase para clases de entre 20 y 200 estudiantes. Una última advertencia: deberá calcular la probabilidad, es decir, la fracción de ensayos en la que dos estudiantes cumplieron el mismo año dividida por el número total de ensayos. Calcular esta fracción es fácil, utilicé "Successes / NumTrials". Sin embargo, ambas variables eran números enteros (por ejemplo, 891/1000), y Python 2.x informa el resultado como un número entero (es decir, 0 o 1) en lugar de una fracción (es decir, 0,891). Para asegurar un resultado de punto flotante, utilicé "(1.0 * éxitos) / NumTrials".


```python
import matplotlib.pyplot as plt
import numpy as np

vx=[]
vy=[]
#Funcion que calcula la probabilidad en 1000 clases
def proba():
    tam=int(input('Escriba el tamaño de la clase que desea '))
    vx.append(tam)
    Num=1000
    con=0
    i=1
    while i<Num:
        a=np.random.randint(1,366,size=tam)
        j=0
        while j<tam:
            h=j+1
            while h<tam:
              if a[j]==a[h]:
                con=con+1
                j=tam
                break
              else:            
                 h=h+1        
            j=j+1
        i=i+1
    x=(con/Num)*1.0
    print('Existe una probabilidad de un ',100*x, '% en la clase de tamaño',tam)
    vy.append(x*100)
proba()
#Se le pregunta al usario si gusta calcular máas probabilidades de acuerdo al tamaño que el guste

confir=input('Desea calcular de nuevo la probabilidad pulse "y" para sí o "n" para finalizar ')
while confir=="y":
  proba()
  confir=input('Desea calcular de nuevo la probabilidad pulse "y" para sí o "n" para finalizar ')

#Gráfico de tamaño de clase vs probabilidad
plt.plot(vx,vy,'o')
plt.grid()
plt.title("Probabilidad de que dos niños cumplan años el mismo día")
plt.ylabel('Probabilidad en %')
plt.xlabel('Tamaño de la clase')
plt.show
```
