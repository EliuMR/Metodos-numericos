To get the best view of a ovie, a person has to sit at a distance x from the srceen such that the viewing angle $\theta$ is maximum. Determine the distance x for wich $\theta$ is maximum.

Utilizando la relación que se nos presenta:

$cos(\theta)=\frac{(x²+5²)+(x²+41²)-36²}{2\sqrt{x²+5²}\sqrt{x²+41²}}$

```python
import sympy
import math
import numpy as np
import matplotlib.pyplot as plt
#Utilizando la libreria para derivar obtenemos la derivada
x=sympy.symbols('x')
cost=((x**2+25)+(x**2+41**2)-36**2)/(2*(x**2+5**2)**(1/2)*(x**2+41**2)**(1/2))
print(sympy.diff(cost,x))

def costeta(x):
    return ((x**2+25)+(x**2+41**2)-36**2)/(2*(x**2+5**2)**(1/2)*(x**2+41**2)**(1/2))
a=np.linspace(0,30,100)
b=costeta(a)
plt.grid()
plt.plot(a,b)
plt.show()
#Dado que cos(0)=1 y el coseno decrese con forme theta incrementa, el máximo ángulo corresponde al menor valor para cos(theta)
```

```python
#Utilizando la derivada la convertimos en una función
#La graficamos para encontrar dónde está su raiz 
def f(x):
    return -0.5*x*(x**2 + 25)**(-1.5)*(x**2 + 1681)**(-0.5)*(2*x**2 + 410) - 0.5*x*(x**2 + 25)**(-0.5)*(x**2 + 1681)**(-1.5)*(2*x**2 + 410) + 2*x*(x**2 + 25)**(-0.5)*(x**2 + 1681)**(-0.5)
l=np.linspace(1,20,100)
m=f(l)
plt.plot(l,m)
plt.grid()
plt.axhline(y=0,color='red')
plt.show()
```

```python
#utilizamos el método de bisección
import numpy as np
def bisectionmethod(h,a,b,maxIter,TOL):
    stop=False
    fa=h(a)
    fb=h(b)
    
    #check starting values
    if(fa*fb>=0):
        print ('error')
    else:
        #Bisection iteration
        for i in range(maxIter):
            c=(a+b)/2
            fc=h(c)
            #print("c=",c,"and f(c)",fc)
            
            #check for convergence
            if(np.fabs(fc)<TOL):
                return c
                #print("Root found at",c)
                stop=True
                break
            #if not conerged, determine end point
            #to replace
            if(fa*fc<0):
                b=c
                fb=fc
            else:
                a=c
                fa=fc
        if stop == False:
            print("Reached maxIter, currently at",c)
root=bisectionmethod(f,12,15,100,0.0001)
print('El valor de x para el cual theta es máximo es: x=',root)
```
